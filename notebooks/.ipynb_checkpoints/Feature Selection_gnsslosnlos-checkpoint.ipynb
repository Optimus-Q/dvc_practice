```python
# Import libraries

import os
import json
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
# datasets

train_df = pd.read_csv("../data/train/GNSS_raw_train.csv")
test_df = pd.read_csv("../data/test/GNSS_raw_test.csv")
```

```python
# selecting the features
_features = ['Satelite_Code', 'Code_L1', 'Phase_L1', 'Doppler_L1',
       'Cnr_L1', 'code_L2', 'Phase_L2', 'Doppler_L2', 'Cnr_L2', 'Label']
```

```python
# training data from selected features
train_data = train_df[_features]
test_data = test_df.copy()
test_data = test_data[_features[:-1]] #remove label from features
```

```python
def DataMapper():
    with open('../data/satelite mapper/satelitecode_mapper.json', 'r') as f:
        map_data = json.load(f)
    return map_data

def DataPipeline(data):
    data_ = data.dropna()
    map_data_satelite_code = DataMapper()
    data_['Satelite_Code'] = data_['Satelite_Code'].map(map_data_satelite_code)
    return data_
```

```python
train_data_updated = DataPipeline(train_data)
test_data_updated = DataPipeline(test_data)
```

```python
train_data_updated.head()
```

```python
test_data_updated.head()
```

```python
# feature selection
```

```python
x = train_data_updated.iloc[:, :-1]
y = train_data_updated.iloc[:, -1]
```

```python
np.unique(y, return_counts=True) #imbalance data
```

```python
xtrain, xtemp, ytrain, ytemp = train_test_split(x, y, train_size = 0.6, random_state = 42, shuffle = True, stratify=y)
```

```python
xtrain.shape, xtemp.shape, ytrain.shape, ytemp.shape
```

```python
np.unique(ytrain, return_counts=True)
```

```python
np.unique(ytemp, return_counts=True)
```

```python
model = RandomForestClassifier().fit(xtrain, ytrain)
```
